## **SIMPLE-IF**


*  simple-if-when condition is true
* another statements when condition is false and when  block is exited



```python
#simple-if-when condition is true
#another statements when condition is false and when if block is exited
age=int(input())
if(age>18):
  print("You can vote,your major")

print(" your a citizen")
```

## **IF-ELSE**

* if-when condition is true
* else-when condition is false





```python
#if-when condition is true
#else-when condition is false
age=int(input())
if(age<18):
  print("Your minor,You cannot vote")#when true
else:
  print("You can vote,your major")#when false
```

## **TERNARY OPERATOR**


*   simply short hand form of if-else
*   List item



```python
age=18
vote_right ="Eligible" if age>=18 else "Not eligible"
print(f"the citizen age is {age} they are {vote_right}")
```

## **IF-ELIF-ELSE LADDER**


*  when there are multiple conditions to be checked  
*  and different task performed based on the condition


```python
citizen_age=int(input())
if(citizen_age<13):
  print("Your are a Child")
elif(citizen_age>=13 and citizen_age<=18):
  print("Your are a teenager")
elif(citizen_age>18 and citizen_age<25):
  print("Your are a Youngster")
elif(citizen_age>=25 and citizen_age<50):
  print("Your are a Middle Ager")
else:
  print("Your are a Senior Citizen")
```

## **NESTED CONDITIONAL STATEMENTS**
* When there a conditions inside conditions
* like if-else or if inside a if or else statements

```python
number=int(input())
if(number>0):
  print(f"{number} is positive and",end=" ")
  if(number%2==0):
    print("is Even")
  else:
    print("is Odd")
else:
  print(f"{number} is Negative")
```

## **MATCH-CASE**
* used when there are multiple condtions that a variable satisfies and different task is performed when each of it is met
* as in other language python doesnt exhibit the fall though mechanism
* so there is no need of break after each case

```python
day=int(input("Enter the number: "))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _: # Default case
        print("Invalid")

```

## **FOR LOOP**


*   executes a block of code repeatedly
*   used when we know where to stop the execution i.e how many times we need to continue repeating the block of code
*  no need to specify the iteration seperately it is automatic
* used to iterate through the sequence data types
* commonly used with range function





```python
for i in range (11):
  print(i) #prints from 0 to n-1(ie 10)
  #if the start in range is not specified it is automatically considered as zero
```

### RANGE


*   is an interable object so you cannot print the items in it directly
* and it works for int only any other gives type error
*   Syntax: range(start,stop,step)

 *   stop is exclusive
 *   step ensure that the mentioned number of elements in the sequence are skipped





```python
print(range(2,10)) # interable so no element is printed
l_r=list(range(2,10)) # using list
print(l_r)
r=range(4,13)
for item in r:# using for
  print(item)
```

## **WHILE LOOP**


*   used when we dont know how many times we need to execute
*   executes block of code untill the condition becomes false
* intialization outside the loop
* iteration should be specified



```python
i=1   #initialize
while i<=10: # runs untill i<=10
  print(i)
  i+=1 #iterate
```

## BREAK
used when we need to get out of the loop when a condition is met

```python
sum=0
while True:
  user=int(input("Enter the numbers: "))
  if(user==0): #when user types 0 exit
    break
  else:
    sum+=user #untill then sum
print(sum)
```

## CONTINUE
used when we need to continue the iteration but skip the rest of the block of code whena certain condition is met

```python
for i in range(11):
  if(i==5):
    continue
  print(i)
```
