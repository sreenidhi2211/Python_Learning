```python
 # access the start element
print(list_int[0]) #positive index
print(list_int[-4])#negative index
# access the last element
print(list_str[-1])
print(list_int[len(list_int)-1])
# access the middle element
print(list_str[(len(list_str)//2)])
print(list_obj[-2])
#acessing the particular element inside a list element
print(list_str[1][2])
```

# MODIFYING THE LIST
* We can modify the list by adding,removing,slicing the list

## LIST SLICING
* we can create sub lists using index from a give list
* syntax: list_name[start,end,step]
* start-start index
* end-end index
* step-no:of elements to skip

```python
list_new=[1,2,3,"string",45.05,23,"A"]
#Accessing entire list
print(list_new[:])
print(list_new[0:])
print(list_new[0:7])
# Accessing elements using negative indexing
print(list_new[-4:-1]) # here the start index -4 ie is 'string; and end is "A"
print(list_new[-1:-2]) #return empty list # since we asked to move form right to left without step
print(list_new[-1:-2:-1]) # now returns last element as step is added
print(list_new[-1:2]) # return empty list # since the index is negative on start and positive at end
list_slice=[2,3,4,5,6,7,8,9]
print(list_slice[0:8:2]) #list of elements at even postion
print(list_slice[-1::-2])# list of elements at odd position from end
#-1 ,-3,-5,-7 indexed elements are printed
```

```python
l_mod=[2,3,4,5,6,7]
l_mod[1:]="ABC" # here we inserted the "ABC " from index 1: so all the elements from 2 are removed and "ABC" is inserted as a individual character
#since string is iterable it is unpacked i.e each character here is treated as individual element
print(l_mod)
l_mod[-1:0]=['as','bs','cs'] # here at -1 the list elements are appended as individuals by pushing the -1 element to the end
print(l_mod)
l_mod[5]=[1,2,3] # the list is insertes at 5 th index
print(l_mod)
```

## INSERTING ELEMENTS INTO LIST
there are two methods used:


*   append():
 1. used to add element at the end of the list.
 2. Only one element can be added at a time
 3. syntax:list_name.append(element)
*   insert():
 1.  used to insert element at a particular index  
 2.  syntax:list_name.insert(index,element)





```python
list_add=[1,2,3,4,5]
# using append
list_add.append(7)
print(list_add)
#you can append another sequence data type in to the list also
list_add.append((8,6,7)) #appending tuple #here the entire tuple is a single element
list_add.append(["s","a","p"]) # appending list
list_add.append({1,2,3,4,5,2,3,2})#appending a set
list_add.append({1:"result",2:"out"}) #appending a dictonary
print(list_add)
```

```python
#using insert
list_1=[1, 2, 3, 4, 5, 7, (8, 6, 7), ['s', 'a', 'p'], {1, 2, 3, 4, 5}, {1: 'result', 2: 'out'}]
list_1.insert(0,"Dear") # at start
print(list_1)
list_1.insert(4,56)
print(list_1)
list_1.insert(7,"sree")
print(list_1)
list_1.insert(-1,"Roolno")# at end
print(list_1)
list_1.insert(-4,345)
print(list_1)
```

## REMOVING ELEMENTS FROM THE LIST
There are three ways through which we remove the elements in the list
*   remove():
1.   removes the first occurance of the element
2.   syntax:list_name.remove(element)
3.   ERROR:if element not present in the list

*   pop():
1.   removes the element at the index and return the popped element
2.   by default it pops last element
3.   syntax:list_name.pop(index)

* del key word:
1. del has the capability of removing entire list elements or even list and parts of list using slicing
2. It doesnt return anything
3. syntax: del list_name[start,stop,step] or del list_name[index] or del list_name





```python
# using remove
l1=["Abc","Def","Hij","Abc"]
l1.remove("Abc")
"""ERROR ALERT:
if element not present in the list:
ValueError: list.remove(x): x not in list
"""
# l1.remove("mno")
print(l1)

#use pop
l1.pop() # by defalut removes last element and returns it
print(l1.pop(1))
print(l1)

#use del keyword
"""
ERROR ALERT:
if the index is out of range error occurs
IndexError: list assignment index out of range
"""
# del l1[2]
del l1[0] # delete an element
print(l1)

del l1 # delete the entire list
# """
# ERROR ALERT:
# once we delete the list and then if we try to access it error occurs
# NameError: name 'l1' is not defined
# """
# print(l1)



```

##Changing list elements using enumerate() and map()

* enumerate(iterable,start):a built in function that iterates through the iterable and assigns a counter to it
    *   returns an enumerate object
    *   the object stores a sequence of tuples with index and the value(index,value)
    *    you need to convert it into list/tuple using list(),tuple()
    *    allows use to track loop index and value simultaneously


* map(function,iterable): a built in iterative function that applies the function to each and every element of the iterable
     *   returns a map object
     *   we convert it into the list/set/tuple

    
* lambda variable_1,varaiable_2: (value to compute),iterable1,iterable2): a short hand function used for computation



    *   same as function but in one line
  






### Enumerate

```python
num_l=[2,5,2,6,8,3,9] # list
enum_l=enumerate(num_l) #returns a enumerate object
print(enum_l)
enum_l=list(enum_l) # to list
print(enum_l)

# track index and value
list_n=[2,1,"peer","leader",3.4]
for i,x in enumerate(list_n):
  print(i,x)

# inplace change in values add 10 to integers
for i,x in enumerate(list_n):
  if type(x)==int:
    list_n[i]+=10
  else:
    continue

print(list_n)


```

### Map

```python
v_list=[2.34,5.63,7.45,9.76,"56"]
# convert v_list into a list of integers
# function to convert to int type
def to_int(i):
  i=int(i)
  return i

i_list=map(to_int,v_list) # stores map object
i_list=list(i_list) #covert into list
print(i_list)

# map and lambda function
# insted of creating a seperate function we use lambda instead
# prepare a new list that is product of corresponding elements of two lists
a=[2,4,6,8]
b=[3,5,7,9]
c=list(map(lambda x,y: x*y,a,b))
print(c)

```
