# Sequence Data Types

```python
st="DEgrfvncm12"
print(type(st))
l=[1,4+5j,True,4.5,"hi"]#list:ordered,mutable,iterable
print(l)
print(type(l))
#tuple:immutable list,non iterable
t=(23,"fed4",True,4+6j)
print(t[2])

print(type(t))
#set:unordered ,unique #so no indexing
s={1,34,"Hi",34,45,True}
print(s)
print(type(s))
#dictionary:key value
d={1:"Hi",2:True}
print(d)
print(type(d))
```

# Type Conversion

```python

#implicit type conversion
# bool->int->float->complex
#no operation should be performed b/w str and bool
a=5
b=10.5
c=a+b # int+float->float
print(c)
print(type(c))
com=15+2j
d=a+com #comp+int->comp
f=b+com #comp+float->comp
print(d)
print(type(d))
print(f)
print(type(f))
name="nidhi"
# Errors:cannot concatenate
'''
g=name+a #str+int
h=name+b #str+float
print(g)
print(type(g))
print(h)
print(type(h))
i=name+com #str+comp
print(i)
'''
#explicit type coversion
a=5
a1=4.56
b=float(a)
print(b)
print(type(b))
c=str(a)
print(c)
print(type(c))
d=complex(a)
print(d)
print(type(d))
e=complex(a1)
print(e)
print(type(e))
f=str(a1)
print(f)
print(type(f))
b1=str(e)
print(b1)
print(type(b1))
t=(1,3,4,2,6,4)
l_t=list(t) # tuple->list
print(l_t)
print(type(l_t))
s_t = set(l_t) # list->set
print(s_t)
print(type(s_t))
```
