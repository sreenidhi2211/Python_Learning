#Fibonacci Sequence
Write a program that prints the first n Fibonacci numbers, where n is input by the user.

```python
series=[]
user=int(input("Enter the number: "))
for i in range(user):
  if i ==0 or i==1:
    series.append(i)
  else:
    series.append((series[i-1]+series[i-2]))

print(f"The first {user} Fibonacci series is:{series}")

```

#Even Odd Sum
Take 10 numbers from user and calculate the sum of even numbers and odd numbers from the list

## Jumping through While - Python

Given a positive integer x, the task is to print the numbers from 1 to x in the order as 1^2, 2^2, 3^2, 4^2, 5^2, ... (in increasing order).
Till the result is less than x

```python
x=15
# using while
i=1
while(i<=x):
  sq=i**2
  if(sq<=x):
    print (sq , end = " ")
    i+=1
  else:
    break
```

## Zero Converter - Python

You are given a number n. The number n can be negative or positive. If n is negative, print numbers from n to 0 by adding 1 to n in the neg function. If positive, print numbers from n-1 to 0 by subtracting 1 from n in the pos function.

```python
# using while loop
def pos(n):
    ## Write the code
    n=n-1
    while(n>=0):
        print(n,end=' ')
        n-=1

def neg(n):
    ##Write the code
    while(n<=0):
        print(n,end=' ')
        n+=1

#using for

def pos(n):
    ## Write the code
    for i in range(n-1,-1,-1):
        print(i,end=" ")

def neg(n):
    ## Write the code
    for i in range(n,1):
        print(i,end=" ")
```

## Right Angle Triangle 2
Given an integer n. Write a program to print the Right angle triangle. The length of the perpendicular and base is n.  

```python
n = int(input())
for i in range(0,n):
  for j in range(0,i+1):
    if(i==0 or i==n-1):# entire row * only when
      print("*",end=" ")
    elif(j== 0 or j==i):# * at ends ends of the row only
      print("*",end=" ")
    else:
      print(" ",end=" ") # rest all blank spaces
  print()
```

## Prime Number

Given a number n, determine whether it is a prime number or not.

```python
#without function
num=int(input())
if num<=1:
  print("false")
elif(num==2):
  print("true")
else:
  is_prime=True
  for i in range(2,num):
    if(num%i==0):
      is_prime=False
      break
  if is_prime:
    print("true")
  else:
    print("false")
```

```python
# with function

#User function Template for python3
n = int(input())

# Your code here
def prime(n):
    if(n==2):
      return "True"
    elif(n<=1):
      return "False"

    else:
        for i in range(2,n):
            if(n%i==0):
                return "False"
                break


        return "True"

print(prime(n))
```

## Next Prime Number
Given an integer n. Write a program to find the first prime number greater than n.



```python
def nextPrime(n):

    # code here to find next prime number
    is_prime = True
    # return next prime number
    for i in range(n+1,500):
        for j in range(2,i):
            if i%j==0:
                is_prime = False
                break
            else:
                is_prime = True
                continue

        if is_prime==True:
            return i
            break
```

* The number must be divisible by five.
If the number is greater than 150, skip it and move to the next.

* If the number is greater than 500, stop the loop entirely.

* Given Input: numbers = [12, 75, 150, 180, 145, 525, 50]

###  Count vowels and consonants in a sentence
